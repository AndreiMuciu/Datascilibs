First, we will take the data from the received csv file and store it in a pandas data frame to make calculations related to these libraries easier. First, we created a new way to sort python libraries, the stars on github not being the most conclusive criterion. The new calculation method takes into account: avg_commits_per_month, recent_contributors, merged_prs, stars, each with its own proportion depending on the importance we give it.

```python
import pandas as pd

df = pd.read_csv('./data.csv')

df['activity_score'] = (
    0.3 * df['avg_commits_per_month'] / df['avg_commits_per_month'].max() +
    0.3 * df['recent_contributors'] / df['recent_contributors'].max() +
    0.2 * df['merged_prs'] / df['merged_prs'].max() +
    0.2 * df['stars'] / df['stars'].max()
)

df_sorted = df.sort_values(by="activity_score", ascending=False)

df_sorted.head(50)
```

We are going to make a table to see more clearly the relationship between popularity (stars) and activity (commits-month). We will also show the top 10 repos with active contributors.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.scatterplot(
    data=df,
    x='stars',
    y='avg_commits_per_month',
    hue='license',
    size='recent_contributors',
    sizes=(20, 200)
    )
plt.xscale('log')
plt.title('Popularity vs. Activity')
plt.show()

top_recent = df.nlargest(10, 'recent_contributors')[['full_name', 'recent_contributors', 'total_contributors']]
top_recent.plot(kind='barh', x='full_name', title='Active Contributors (Last 6 Months)')
```

To measure how much attention the project receives per contributor, we will create a new attribute: "engagement-effort", which will reflect the ratio of the project's stars to the total number of contributors. This way, it can be shown how much impact some projects have had with a small number a contributors.

```python
df['engagement_to_effort'] = (
    df['stars'] / df['total_contributors']
)

df_sorted = df.sort_values(by="engagement_to_effort", ascending=False)

df_sorted.head(50)
```

Also, to be able to differentiate passively popular projects ( just many stars ) from those that have active interest ( many forks ), we will create the fork-to-star attribute which should show us exactly that.

```python
df['fork_to_star'] = (
    df['forks'] / df['stars']
)

df_sorted = df.sort_values(by="fork_to_star", ascending=False)

df_sorted.head(50)
```

To measure the request acceptance rate, we will use PR-Merge-Rate attribute, which will show us how difficult it is for someone who wants to contribute to projects and what chances are of his pull request being accepted.

```python
df['PR_Merge_Rate'] = (
    df['merged_prs'] / (df['merged_prs'] + df['closed_prs']) * 100
)

df_sorted = df.sort_values(by="PR_Merge_Rate", ascending=False)

df_sorted.head(50)
```

Commit Momentum is a compound metric that captures the current energy and development pace of an open-source project. It combines the average monthly commit frequency with the number of active contributors in a recent time window. Unlike simple commit counts, Commit Momentum highlights both consistency and collaborative activity. It helps differentiate between projects that are being maintained by a single person at high pace vs. those sustained by a broader, active community.

```python
df['Commit_Momentum'] = (
    df['avg_commits_per_month'] * df['recent_contributors']
)

df_sorted = df.sort_values(by="Commit_Momentum", ascending=False)

df_sorted.head(50)
```

Open Issues per Contributor is a metric that measures the current maintenance load of an open-source project by comparing the number of unresolved issues to the total number of contributors. Projects with a high ratio may struggle to keep up with bug reports, feature requests, or community feedback, potentially affecting user satisfaction and long-term sustainability. On the other hand, a low ratio often reflects good project health and an adequately resourced or well-organized contributor team.

```python
import numpy as np
```

```python
df['Open_Issues_per_Contributor'] = (
      df['open_issues'] / df['recent_contributors']
)
df['Open_Issues_per_Contributor'] = (df['open_issues'] / df['recent_contributors']).replace([np.inf, -np.inf], 0).fillna(0)

df_sorted = df.sort_values(by="Open_Issues_per_Contributor", ascending=False)

df_sorted.head(50)
```

Releases per Star is a metric that measures the ratio of releases to the popularity of the project as reflected by the number of stars it has on GitHub. This metric helps indicate how frequently a project releases new versions relative to how popular it is. A higher ratio suggests that the project is frequently updated regardless of its popularity, while a lower ratio may suggest that the project is either very stable or not updated very often, despite its popularity.

```python
from datetime import datetime

df['Releases_per_Star'] = df['number_of_releases'] / df['stars']

df_sorted = df.sort_values(by="Releases_per_Star", ascending=False)

df_sorted.head(50)
```

```python
from scipy.stats import linregress

def calculate_contributor_growth(repo):
    try:
        stats = repo.get_stats_contributors()
        time.sleep(1)
        
        if not stats:
            return None

        weekly_totals = [0] * 52  # 52 sptmni
        for contributor in stats:
            for i, week in enumerate(contributor.weeks[-52:]):
                weekly_totals[i] += week.c  # .c = commits

        # Convertim commituri n numr de contribuitori activi pe sptmn
        active_contributors_per_week = [1 if count > 0 else 0 for count in weekly_totals]

        # Calculm trendul prin regresie liniar (x = sptmna, y = contribuitori)
        x = list(range(len(active_contributors_per_week)))
        y = active_contributors_per_week

        slope, _, _, _, _ = linregress(x, y)

        return slope  # pozitiv = cretere, negativ = scdere

    except Exception as e:
        print(f"  Eroare contributor trend la {repo.full_name}: {e}")
        return None
```

```python
from itertools import islice

def analyze_issues(repo):
    try:
        closed_issues = 0
        opened_issues = 0
        user_closure_count = {}
        label_resolution_times = {}
        bug_issues = 0
        invalid_issues = 0
        reopened_bugs = 0
        durations = []

        # Ia doar ultimele 10 issue-uri (excludem PR-urile mai jos)
        issues = islice(repo.get_issues(state='all', sort='created', direction='desc'), 20)
        time.sleep(1)

        for issue in issues:
            if issue.pull_request is not None:
                continue  # Skip PRs

            labels = [label.name.lower() for label in issue.labels]

            if issue.state == "closed":
                closed_issues += 1
                if issue.closed_at and issue.created_at:
                    duration = (issue.closed_at - issue.created_at).total_seconds()
                    durations.append(duration)

                    for label in labels:
                        label_resolution_times.setdefault(label, []).append(duration)

                if issue.closed_by:
                    user = issue.closed_by.login
                    user_closure_count[user] = user_closure_count.get(user, 0) + 1

                if 'bug' in labels and issue.reactions.get("+1") > 0:
                    reopened_bugs += 1

            elif issue.state == "open":
                opened_issues += 1

            if 'bug' in labels:
                bug_issues += 1
            if 'invalid' in labels:
                invalid_issues += 1

        top_solver = max(user_closure_count.items(), key=lambda x: x[1])[0] if user_closure_count else None

        avg_times_by_label = {
            label: sum(times) / len(times)
            for label, times in label_resolution_times.items() if times
        }
        fastest_label = min(avg_times_by_label.items(), key=lambda x: x[1])[0] if avg_times_by_label else None

        avg_resolution_speed = sum(durations) / len(durations) / 3600 if durations else None

        w1, w2, w3 = 3.0, 1.5, 1.0
        buggy_level = w1 * bug_issues + w2 * invalid_issues + w3 * (reopened_bugs / (bug_issues + 1))

        return {
            "closed_issues": closed_issues,
            "opened_issues": opened_issues,
            "issues_top_solver": top_solver,
            "issues_fastest_solved_label": fastest_label,
            "issue_resolution_speed": avg_resolution_speed,
            "repo_buggy_lvl": buggy_level
        }

    except Exception as e:
        print(f"  Eroare la analiz issue-uri: {e}")
        return {
            "closed_issues": None,
            "opened_issues": None,
            "issues_top_solver": None,
            "issues_fastest_solved_label": None,
            "issue_resolution_speed": None,
            "repo_buggy_lvl": None
        }
```

```python
from github import Github
from dotenv import load_dotenv
import os
from urllib.parse import urlparse
from datetime import datetime
import pandas as pd
import time

load_dotenv()
token = os.getenv("TOKEN")
gh = Github(login_or_token=token)

test_df = df.head(10).copy()

def extract_owner_repo(url):
    path_parts = urlparse(url).path.strip("/").split("/")
    if len(path_parts) >= 2:
        return path_parts[0], path_parts[1]
    return None, None

avg_durations = []

contributor_growth_trends = []

for index, row in df.iterrows():
    url = row['url']
    owner, repo_name = extract_owner_repo(url)

    if owner and repo_name:
        try:
            repo = gh.get_repo(f"{owner}/{repo_name}")
            time.sleep(1)

            # PR merge velocity
            pulls = repo.get_pulls(state='closed', sort='updated')
            durations = []
            for i, pr in enumerate(pulls):
                if i >= 10:
                    break
                if pr.closed_at and pr.created_at:
                    durations.append((pr.closed_at - pr.created_at).total_seconds())
            avg_seconds = sum(durations) / len(durations) if durations else None
            avg_durations.append(avg_seconds / 3600 if avg_seconds else None)

            # Contributor growth trend
            slope = calculate_contributor_growth(repo)
            contributor_growth_trends.append(slope)

        except Exception as e:
            print(f"  Eroare la {owner}/{repo_name}: {e}")
            avg_durations.append(None)
            contributor_growth_trends.append(None)
    else:
        print(f"  Nu am putut extrage repo din URL: {url}")
        avg_durations.append(None)
        contributor_growth_trends.append(None)

df["pr_merge_velocity"] = avg_durations
df["contributor_growth_trend"] = contributor_growth_trends

df.head(10)

```

```python
closed_issues_list = []
opened_issues_list = []
top_solvers = []
fastest_labels = []
issue_speeds = []
bugginess_levels = []

for index, row in df.iterrows():
    url = row['url']
    owner, repo_name = extract_owner_repo(url)

    if owner and repo_name:
        try:
            repo = gh.get_repo(f"{owner}/{repo_name}")
            time.sleep(1)

            issue_metrics = analyze_issues(repo)

            closed_issues_list.append(issue_metrics["closed_issues"])
            opened_issues_list.append(issue_metrics["opened_issues"])
            top_solvers.append(issue_metrics["issues_top_solver"])
            fastest_labels.append(issue_metrics["issues_fastest_solved_label"])
            issue_speeds.append(issue_metrics["issue_resolution_speed"])
            bugginess_levels.append(issue_metrics["repo_buggy_lvl"])

        except Exception as e:
            print(f"  Eroare la {owner}/{repo_name}: {e}")
            avg_durations.append(None)
            contributor_growth_trends.append(None)
    else:
        print(f"  Nu am putut extrage repo din URL: {url}")
        avg_durations.append(None)
        contributor_growth_trends.append(None)
        
df["closed_issues"] = closed_issues_list
df["opened_issues"] = opened_issues_list
df["issues_top_solver"] = top_solvers
df["issues_fastest_solved_label"] = fastest_labels
df["issue_resolution_speed"] = issue_speeds
df["repo_buggy_lvl"] = bugginess_levels
```
